```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression #using liaberery
from sklearn.model_selection import train_test_split #using this for spliting data
from sklearn.metrics import accuracy_score
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("loan_synthetic.csv")
print(df)

#target_the_valuedf
x=df[["age","income","employment_years","credit_score","loan_amount","loan_term_months","dependents","existing_debt","num_of_credit_lines","previous_defaults","months_since_last_default","debt_to_income","loan_to_income","interest_rate","approval_prob"]]

x=x.fillna(x.median())
print(df.isnull().sum())

y=df[["approved"]]

#prediction data

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)

user_input=np.array([[22, 50000, 2.0, 600, 15000, 12, 0, 2000, 1, 0,0,0.191,0.228,0.122,0.503]])
user_pred=model.predict(user_input)
print("posbility of loan:  ",(int(user_pred)))

accurcy_score=accuracy_score(y_test,pred)
print("accurcy score:  ",accurcy_score)


# age=int(input("Enter he age:"))
# income=float(input("Enter the income:"))
# employment_years=float(input("Enter the employment_years:"))
# credit_score=float(input("Enter the credit_score:"))
# loan_amount=float(input("Enter the total loan on it:"))
# loan_term_months=int((input("Enter the months that spend after taking loan:")))
# dependents=int((input("Enter how many member depend on it:")))
# existing_debt=float(input("Enter the existing_debt:"))
# num_of_credit_lines=int(input("Enter the num_of_credit_lines:"))
# previous_defaults=int((input("Enter the previous_defaults:")))
# months_since_last_default=float(input("Enter the months_since_last_default:"))
# debt_to_income=float(input("Enter the debt_to_income"))
# loan_to_income=float(input("Enter the loan to income:"))
# intrest_rate=float(input("Enter the intrest_rate:"))
# approval_prob=float(input("Enter the approval probability:"))

# user_input=[[age,income,employment_years,credit_score,loan_amount,loan_term_months,dependents,existing_debt,num_of_credit_lines,previous_defaults,months_since_last_default,debt_to_income,loan_to_income,interest_rate,approval_prob]]
# prediction=model.predict(user_input)
```
