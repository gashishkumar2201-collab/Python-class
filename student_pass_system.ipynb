```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression #using liaberery
from sklearn.model_selection import train_test_split #using this for spliting data
from sklearn.metrics import accuracy_score
#import numpy as np
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("students_1000.csv")
print(df)
x=df[["study_hours","sleep_hours",]]#factors provide
y=df[["pass"]]
#print(x)
#print(y)
#plt.title("student data")
#plt.xlabel("student sleeping_houre")
#plt.ylabel("pass")
#plt.plot(x,y)
#plt.show()
#plt.title("student data")
#training of data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LogisticRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)
#print(pred)
#user data prediction
#user_input=np.array([[6.99,9.1]])
student_study_hour=int(input("student_study_hour:  "))
student_sleeping_hour=int(input("student_sleeping_hour:  "))
user_list=[[student_study_hour,student_sleeping_hour]]

user_pred=model.predict(user_list)
if student_sleeping_hour<=10:
  if user_pred[0]==1:
   print("you are pass")
  else:
   print("you need to do hard work")
else:
  print("you need to reduce sleeping hours")
print("student result:  ",(int(user_pred)))
if user_pred[0] == 1:
  print("passed")
else:
  print("failed")
accurcy_score=accuracy_score(y_test,pred)
print("accurcy score",accurcy_score)
```
