```python
# ==========================
# STUDENT MANAGEMENT SYSTEM
# ==========================

# ---- Load existing data ----
def load_students(filename="students.txt"):
    students = []
    try:
        with open(filename, "r") as f:
            for line in f:
                name, age, marks = line.strip().split(",")
                students.append({"name": name, "age": int(age), "marks": float(marks)})
    except FileNotFoundError:
        print("⚠️ No previous data found. Starting fresh.")
    return students


# ---- Save data to file ----
def save_students(students, filename="students.txt"):
    try:
        with open(filename, "w") as f:
            for s in students:
                f.write(f"{s['name']},{s['age']},{s['marks']}\n")
        print("💾 Data saved successfully.")
    except Exception as e:
        print("❌ Error while saving:", e)


# ---- Add student ----
def add_student(students):
    try:
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        marks = float(input("Enter marks: "))
        student = {"name": name, "age": age, "marks": marks}
        students.append(student)
        print("✅ Student added successfully!")
    except Exception as e:
        print("❌ Error:", e)


# ---- View all students ----
def view_students(students):
    if not students:
        print("No students available.")
    else:
        for i, student in enumerate(students, start=1):
            print(f"{i}. {student['name']} | Age: {student['age']} | Marks: {student['marks']}")


# ---- Search student ----
def search_student(students):
    name = input("Enter name to search: ").lower()
    found = False
    for student in students:
        if name in student["name"].lower():
            print(f"Found: {student['name']} | Age: {student['age']} | Marks: {student['marks']}")
            found = True
    if not found:
        print("❌ Student not found.")


# ---- Update student ----
def update_student(students):
    name = input("Enter name to update: ").lower()
    for student in students:
        if name == student["name"].lower():
            new_marks = float(input("Enter new marks: "))
            student["marks"] = new_marks
            print("✅ Marks updated successfully!")
            return
    print("❌ Student not found.")


# ---- Delete student ----
def delete_student(students):
    name = input("Enter name to delete: ").lower()
    for student in students:
        if name == student["name"].lower():
            students.remove(student)
            print("✅ Student deleted successfully!")
            return
    print("❌ Student not found.")


# ---- Show statistics (tuple example) ----
def statistics(students):
    if not students:
        print("No students available.")
        return
    marks = [s["marks"] for s in students]
    highest = max(marks)
    lowest = min(marks)
    avg = sum(marks) / len(marks)
    stats = (highest, lowest, avg)
    print(f"📊 Highest: {stats[0]}, Lowest: {stats[1]}, Average: {stats[2]:.2f}")


# ---- Menu ----
def menu():
    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Show Statistics")
    print("7. Save & Exit")


# ---- Main Program ----
def main():
    students = load_students()

    while True:
        menu()
        choice = input("Enter choice: ")

        if choice == "1":
            add_student(students)
        elif choice == "2":
            view_students(students)
        elif choice == "3":
            search_student(students)
        elif choice == "4":
            update_student(students)
        elif choice == "5":
            delete_student(students)
        elif choice == "6":
            statistics(students)
        elif choice == "7":
            save_students(students)
            print("Exiting... Goodbye! 👋")
            break
        else:
            print("❌ Invalid choice. Try again.")


# ---- Run the program ----
if __name__ == "__main__":
    main()
```

```python
# ==========================
# STUDENT MANAGEMENT SYSTEM
# ==========================

# ---- Load existing data ----
def load_students(filename="students.txt"):
    students = []
    try:
        with open(filename, "r") as f:
            for line in f:
                data = line.strip().split(",")
                name = data[0]
                age = int(data[1])
                marks = {
                    "hindi": float(data[2]),
                    "english": float(data[3]),
                    "math": float(data[4]),
                    "science": float(data[5]),
                    "social": float(data[6])
                }
                students.append({"name": name, "age": age, "marks": marks})
    except FileNotFoundError:
        print("⚠️ No previous data found. Starting fresh.")
    return students


# ---- Save data to file ----
def save_students(students, filename="students.txt"):
    try:
        with open(filename, "w") as f:
            for s in students:
                m = s["marks"]
                f.write(f"{s['name']},{s['age']},{m['hindi']},{m['english']},{m['math']},{m['science']},{m['social']}\n")
        print("💾 Data saved successfully.")
    except Exception as e:
        print("❌ Error while saving:", e)


# ---- Add student ----
def add_student(students):
    try:
        name = input("Enter name: ")
        age = int(input("Enter age: "))
        print("Enter marks for 5 subjects:")
        hindi = float(input("Hindi: "))
        english = float(input("English: "))
        math = float(input("Math: "))
        science = float(input("Science: "))
        social = float(input("Social Studies: "))

        marks = {
            "hindi": hindi,
            "english": english,
            "math": math,
            "science": science,
            "social": social
        }
        students.append({"name": name, "age": age, "marks": marks})
        print("✅ Student added successfully!")
    except Exception as e:
        print("❌ Error:", e)


# ---- View all students ----
def view_students(students):
    if not students:
        print("No students available.")
    else:
        for i, s in enumerate(students, start=1):
            m = s["marks"]
            total = sum(m.values())
            avg = total / 5
            print(f"{i}. {s['name']} | Age: {s['age']} | "
                  f"H:{m['hindi']} E:{m['english']} M:{m['math']} S:{m['science']} SS:{m['social']} | "
                  f"Total: {total} | Avg: {avg:.2f}")


# ---- Search student ----
def search_student(students):
    name = input("Enter name to search: ").lower()
    found = False
    for s in students:
        if name in s["name"].lower():
            m = s["marks"]
            total = sum(m.values())
            avg = total / 5
            print(f"Found: {s['name']} | Age: {s['age']} | Marks: {m} | Total: {total} | Avg: {avg:.2f}")
            found = True
    if not found:
        print("❌ Student not found.")


# ---- Update student marks ----
def update_student(students):
    name = input("Enter name to update: ").lower()
    for s in students:
        if name == s["name"].lower():
            print("Enter new marks (leave blank to keep old):")
            for subject in s["marks"]:
                new_mark = input(f"{subject.capitalize()} ({s['marks'][subject]}): ")
                if new_mark.strip() != "":
                    s["marks"][subject] = float(new_mark)
            print("✅ Marks updated successfully!")
            return
    print("❌ Student not found.")


# ---- Delete student ----
def delete_student(students):
    name = input("Enter name to delete: ").lower()
    for s in students:
        if name == s["name"].lower():
            students.remove(s)
            print("✅ Student deleted successfully!")
            return
    print("❌ Student not found.")


# ---- Show statistics (tuple example) ----
def statistics(students):
    if not students:
        print("No students available.")
        return

    all_totals = [sum(s["marks"].values()) for s in students]
    highest = max(all_totals)
    lowest = min(all_totals)
    avg = sum(all_totals) / len(all_totals)
    stats = (highest, lowest, avg)
    print(f"📊 Highest Total: {stats[0]}, Lowest Total: {stats[1]}, Average Total: {stats[2]:.2f}")


# ---- Menu ----
def menu():
    print("\n===== Student Management System =====")
    print("1. Add Student")
    print("2. View All Students")
    print("3. Search Student")
    print("4. Update Student")
    print("5. Delete Student")
    print("6. Show Statistics")
    print("7. Save & Exit")


# ---- Main Program ----
def main():
    students = load_students()

    while True:
        menu()
        choice = input("Enter choice: ")

        if choice == "1":
            add_student(students)
        elif choice == "2":
            view_students(students)
        elif choice == "3":
            search_student(students)
        elif choice == "4":
            update_student(students)
        elif choice == "5":
            delete_student(students)
        elif choice == "6":
            statistics(students)
        elif choice == "7":
            save_students(students)
            print("Exiting... Goodbye! 👋")
            break
        else:
            print("❌ Invalid choice. Try again.")


if __name__ == "__main__":
    main()
```
