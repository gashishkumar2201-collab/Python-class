```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression #using liaberery
from sklearn.model_selection import train_test_split #using this for spliting data
from sklearn.metrics import accuracy_score
import numpy as np
import warnings
warnings.filterwarnings('ignore')
df=pd.read_csv("indexData.csv")
print(df)
x=df[["Open","High","Low","Adj Close","Volume"]]

x=x.fillna(x.median())

y=df[["Close"]]
y=y.fillna(y.median())

print(x.isnull().sum())

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
model=LinearRegression()
model.fit(x_train,y_train)
pred=model.predict(x_test)

user_input=np.array([[531.119995,531.119995,531.119995,531.119995,0]])
user_pred=model.predict(user_input)
print("posbility of  stock:  ",(int(user_pred)))

#accurcy test

```
