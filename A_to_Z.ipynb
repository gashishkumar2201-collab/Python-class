The Fibonacci sequence is a series of numbers where each number is the sum of the two preceding ones, usually starting with 0 and 1. We can generate this sequence using a `for` loop by keeping track of the previous two numbers in the sequence.

```python
n_terms = int(input(""))

a, b = 0, 1

print("Fibonacci sequence:")
for _ in range(n_terms):
    print(a)
    nth = a + b
    a = b
    b = nth
```
